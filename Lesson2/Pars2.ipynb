```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import lxml
from tqdm import tqdm
```

# def

```python
PATH = 'https://roscontrol.com/category/produkti/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.116 Safari/537.36'}
```

```python
url = 'https://roscontrol.com/category/produkti/molochnie_produkti/moloko/?page=1'
def bs_re(url, pars = 'lxml', headers = headers):
  requests_ = requests.get(url, headers = headers)
  if requests_.status_code == 200:
    print('Good 200')
    soup = BeautifulSoup(requests_.text, pars)
  else:
    print(requests_.status_code)
    # break
  return soup
```

```python
# Как правильно справиться с перелистыванием страниц не разобрался. А использование try-except не 
# помогло так как при добавлении несуществующей страницы сайт кидает на первую, таким
# образом попадаешь в бесконечный цикл. Поэтому распарсил число страниц

def number_pages(url):
  soup = bs_re(url)
  roots = soup.find_all('div', attrs = {'class': "pagination util-margin-clear"})
  len_ = len([root.text.split() for root in roots][0]) - 2
  if len_ < 0: 
    page_namber = 1 
  else: 
    page_namber = len_
  # [root.text.split() for root in roots][0]
  print(page_namber)
  return page_namber

# тестируем подвид продукта с числом страниц 5
url = 'https://roscontrol.com/category/produkti/molochnie_produkti/moloko/?page=1'  
number_pages(url)

# тестируем подвид продукта со страницей 1 (должно поулчится 1)
url = 'https://roscontrol.com/category/produkti/molochnie_produkti/prostokvasha/'
number_pages(url)
```

# Go

```python
#  ПО заданию нужно распарсить только продукты
# Структура сайта такая: Продукты/Вид продукта(Молочная продукция)/Подвид товара(Молоко)/Конкретный продукт(Ашан 3,5%)
# Оценки продукта и ссылку можно было вытащить уже на стадии "подвид товара", но я решил дойти до конца

url = 'https://roscontrol.com/category/produkti/'
soup = bs_re(url, pars = 'html.parser')
```

```python
# Словарик чтобы можно было по ссылке вытащить название на русском 
url_product = []
selector = '.catalog__category-name'
keys = [i.text.replace("", '') for i in soup.select(selector=selector)]

for i in soup.select('.testlab-category a'):
  url_one = i.get('href')
  url_product.append(url_one.split('/')[-2])

dict_prod = {key: value for key, value in zip(keys, url_product)}
dict_prod
```

```python
# Формирую ссылки на виды товаров у продуктов
n = 0
list_url = []
for i in dict_prod.values():
  requests_ = requests.get(('https://roscontrol.com/category/produkti/' + i))
  print(('https://roscontrol.com/category/produkti/' + i))
  list_url.append('https://roscontrol.com/category/produkti/' + i)
```

```python
#  Всего видов:
len(list_url)
```

```python
# ЧТобы не запутаться в циклах всегда создаю список куда записывая все нужные ссылки. Тут создаю ссылки на подвиды каждого вида продуктов
url_product = []
select3 = '.main-container .catalog__category-item'
for url in tqdm(list_url):
  soup = bs_re(url)
  for i in soup.select(select3):
    url_one = i.get('href')
    url_product.append(url_one)
    print(url_one)
  url_product
```

```python
%%time

# Для удобства вывожу: ссылка на подвид продукта с указанием страницы и перечень товаров входящую в эту страницу
# И все ссылки собираю в uniq_url_product

uniq_url_product = []
for i in tqdm(url_product):
  url = 'https://roscontrol.com' + i + '?page=1'
  page_numbers = number_pages(url)
  for page_number in range(1, page_numbers + 1, 1):
    soup = bs_re('https://roscontrol.com' + i  + '?page=' + str(page_number))
    print('https://roscontrol.com' + i  + '?page=' + str(page_number))
    select3 = '.main-container .block-product-catalog__item'
    for ii in soup.select(select3):
      url_one = ii.get('href')
      uniq_url_product.append(url_one)
      print(url_one)
```

```python
# всего продуктов

len(uniq_url_product)
```

```python
# На колабе код бежал час

dict_ = {}
for i_uup in tqdm(uniq_url_product):
  url = 'https://roscontrol.com/' + str(i_uup)
  soup1 = bs_re(url)
  dict_[soup1.find('h1').text] = {}
  dict_[soup1.find('h1').text]['scores'] = [(scores.text.split()[0]) for scores in soup1.find_all(['div'], attrs= {'class': ['rate-item__value']})]
  try:
    dict_[soup1.find('h1').text]['common_scores'] = int(soup1.find(['div'], attrs= {'class': ['product__single-rev-total']}).text.split()[0])# [int(scores.text.split()[0]) for scores in soup1.find_all(['div'], attrs= {'class': ['product__single-rev-total']})]
  except:
    # продукты из черного списка или у которых нет оценки
    dict_[soup1.find('h1').text]['common_scores'] = 0
  try:
    dict_[soup1.find('h1').text]['info'] = soup1.find('div', attrs={'class':'p-data'}).text.replace('\n', '')
  except:
    # продукты у которых нет описания
    dict_[soup1.find('h1').text]['info'] = "no info"
  dict_[soup1.find('h1').text]['link'] = url
```
