# Task2

Напишите запрос для поиска продуктов с рейтингом не ниже введенного или качеством не ниже введенного (то есть цифра вводится одна, а запрос проверяет оба поля)

```python
def get_score_quality(colection, score):
  result = colection.find(
      {'$or': [
              {'common_scores' : {'$gt': score}},
              {'Качество': {'gt': score}}
              ]})
  return result
```

```python
# у меня парсинг длится час, поэтому воспользовался уже готовыми файлами
with open('E:/Host/Data/Day1310/df_json.json', 'r', encoding='utf-8') as f:
  json_df = json.load(f)
json_df
```

```python
client = MongoClient('localhost', 27017)

db = client['test_database']

db.test_collection.drop()

collection = db.test_collection

collection.insert_many(json_df)

data = get_score_quality(collection, 80)

for document in data:
    pprint(document)
```
